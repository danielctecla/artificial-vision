```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data points
c1 = np.array([[0,1,2,3],
              [0,2,3,3]])
c2 = np.array([[5,4,5,6],
              [4,5,10,7]])
c3 = np.array([[9,10,7,8],
               [10,11,6,7]])

xVect = [6,5]
```

```python
def getCentroid(data: np.ndarray) -> list[float]:
  n, m = data.shape

  sums = [0] * n
  for i in range(n):
      for j in range(m):
          sums[i] += data[i, j]

  centroid = [sums[i] / m for i in range(n)]
  return centroid

def getDistance(ctd: list[float], vect: list[float]) -> float:
  return np.sqrt((ctd[0] - vect[0])**2 + (ctd[1] - vect[1])**2)
```

```python
fig, ax = plt.subplots()

ax.scatter(c1[0, :], c1[1, :], c="g")
ax.scatter(c2[0, :], c2[1, :], c="b")
ax.scatter(c3[0, :], c3[1, :], c="r")

ctd1 = getCentroid(c1)
ctd2 = getCentroid(c2)
ctd3 = getCentroid(c3)
ax.scatter(ctd1[0], ctd1[1], c="g", marker="x", label="Centroid 1")
ax.scatter(ctd2[0], ctd2[1], c="b", marker="x", label="Centroid 2")
ax.scatter(ctd3[0], ctd3[1], c="r", marker="x", label="Centroid 3")

ax.scatter(xVect[0], xVect[1], c="purple", label=r"$\vec{x}$")

ax.set_title("Clustering")
ax.legend()
ax.grid(True)
plt.show()
```

```python
print("Distancia del vector x con respecto a los centroides")
print(f"C1: {getDistance(ctd1,xVect)}")
print(f"C2: {getDistance(ctd2,xVect)}")
print(f"C3: {getDistance(ctd3,xVect)}")
fig, ax = plt.subplots()

ax.scatter(c1[0, :], c1[1, :], c="g")
ax.scatter(c2[0, :], c2[1, :], c="b")
ax.scatter(c3[0, :], c3[1, :], c="r")

ax.scatter(ctd1[0], ctd1[1], c="g", marker="x", label="Centroid 1")
ax.scatter(ctd2[0], ctd2[1], c="b", marker="x", label="Centroid 2")
ax.scatter(ctd3[0], ctd3[1], c="r", marker="x", label="Centroid 3")

ax.plot([ctd2[0],xVect[0]], [ctd2[1],xVect[1]], c="black")
ax.scatter(xVect[0], xVect[1], c="b", label=r"$\vec{x}$")
ax.text(5.0,5.5,r"$1.8$", fontsize=7)

ax.set_title("Clustering")
ax.legend()
ax.grid(True)
plt.show()
```
